```python
import pandas as pd
import pymssql
from sqlalchemy import create_engine
import get_secrets
```

```python
secrets = get_secrets.get_secrets('.secrets')

server = secrets[4]
database = secrets[5]
username = secrets[6]
password = secrets[7]

string = "mssql+pymssql://" + username + ":" + password + "@" + server + "/" + database 
conn = create_engine(string).connect()
```

```python
df = pd.read_sql(
    "select * from information_schema.tables",
    conn
)
df
```

```python
df = pd.read_sql(
    """
    select * from sf1 
    where ticker='AAPL' and dimension='ARY'
    order by reportperiod
    """, 
    conn
)
df
```

```python
df = pd.read_sql(
    """
    select * from sf1 
    where ticker in ('AAPL', 'CVX') and dimension='ARQ' and reportperiod>='2020-01-01'
    order by ticker, reportperiod
    """, 
    conn
)
df
```

```python
df = pd.read_sql(
    """
    select * from sep where ticker='AAPL' 
    order by date
    """,
    conn
)
df
```
